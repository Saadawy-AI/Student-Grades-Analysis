```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ÿ•ÿπÿØÿßÿØ ÿßŸÑÿπÿ±ÿ∂
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)

# ÿ™ÿ≠ŸÖŸäŸÑ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
df = pd.read_csv("Student Grades.csv")

# ----------- ÿ•ÿ≠ÿµÿßÿ°ÿßÿ™ ÿπÿßŸÖÿ© -----------

print("\nüìä ÿ•ÿ≠ÿµÿßÿ¶Ÿäÿßÿ™ ŸàÿµŸÅŸäÿ©:\n", df.describe())

# ŸÇÿßÿ¶ŸÖÿ© ÿßŸÑŸÖŸàÿßÿØ
subjects = ["Math", "Physics", "Chemistry", "English", "Biology", "History"]

# ----------- Histogram ŸÑŸÉŸÑ ŸÖÿßÿØÿ© -----------
for subject in subjects:
    plt.figure()
    sns.histplot(df[subject], kde=True, bins=10, color='skyblue')
    plt.title(f'Distribution of Grades in {subject}')
    plt.xlabel('Grade')
    plt.ylabel('Number of Students')
    plt.grid(True)
    plt.show()

# ----------- Boxplot ŸÖŸÇÿßÿ±ŸÜÿ© ÿ®ŸäŸÜ ÿßŸÑŸÖŸàÿßÿØ -----------
plt.figure()
sns.boxplot(data=df[subjects])
plt.title("Boxplot of Grades for All Subjects")
plt.xticks(rotation=45)
plt.show()

# ----------- Heatmap ŸÑÿπŸÑÿßŸÇÿ© ÿßŸÑŸÖŸàÿßÿØ ÿ®ÿ®ÿπÿ∂Ÿáÿß -----------
plt.figure()
corr_matrix = df[subjects].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation between Subjects")
plt.show()

# ----------- Barplot ÿ™Ÿàÿ≤Ÿäÿπ ÿßŸÑÿ™ŸÇÿØŸäÿ±ÿßÿ™ ÿßŸÑÿ≠ÿ±ŸÅŸäÿ© -----------
plt.figure()
sns.countplot(x='Grade Letter', data=df, order=['A', 'B', 'C', 'D', 'E', 'F'], palette='Set2')
plt.title("Distribution of Grade Letters")
plt.xlabel("Grade Letter")
plt.ylabel("Number of Students")
plt.show()

# ----------- Pie chart ŸÑŸÜÿ≥ÿ® ÿßŸÑÿ™ŸÇÿØŸäÿ±ÿßÿ™ ÿßŸÑÿ≠ÿ±ŸÅŸäÿ© -----------
plt.figure()
grade_counts = df['Grade Letter'].value_counts().sort_index()
plt.pie(grade_counts, labels=grade_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("pastel"))
plt.title("Grade Letter Distribution")
plt.axis('equal')
plt.show()

# ----------- Barplot ŸÑÿπÿØÿØ ÿßŸÑÿ≥ÿßÿπÿßÿ™ ÿßŸÑŸÖÿ≥ŸÖŸàÿ≠ÿ© ÿ≠ÿ≥ÿ® ÿßŸÑÿ™ŸÇÿØŸäÿ± -----------
plt.figure()
sns.boxplot(x='Grade Letter', y='Allowed Hours Next Semester', data=df, order=['A', 'B', 'C', 'D', 'E', 'F'])
plt.title("Allowed Hours by Grade Letter")
plt.show()

# ----------- Pairplot ŸÑÿπŸÑÿßŸÇÿßÿ™ ÿßŸÑŸÖŸàÿßÿØ ÿ®ÿ®ÿπÿ∂Ÿáÿß -----------
sns.pairplot(df[subjects + ['Final Grade']])
plt.suptitle("Pairwise Relationships Between Subjects", y=1.02)
plt.show()

# ----------- ÿÆÿ∑ ÿ®ŸäÿßŸÜŸä ŸÑŸÖŸÇÿßÿ±ŸÜÿ© ÿØÿ±ÿ¨ÿßÿ™ ŸÉŸÑ ÿ∑ÿßŸÑÿ® ÿπÿ®ÿ± ÿßŸÑŸÖŸàÿßÿØ -----------
plt.figure(figsize=(14, 6))
for i, row in df.iterrows():
    plt.plot(subjects, row[subjects], marker='o', label=row['Name'])
plt.title("Student Grades Across All Subjects")
plt.xlabel("Subjects")
plt.ylabel("Grades")
plt.legend(loc='upper right', bbox_to_anchor=(1.15, 1))
plt.tight_layout()
plt.show()
```
